```python
#Task4

img = io.imread("images/01.jpg",1)

# Self defined sharpen kernel
kernel = np.array([[-1,-1,-1], 
                   [-1, 9,-1],
                   [-1,-1,-1]])

#function call: scratch built
image_sharpen = util.Self_Convolution_Operation(img,kernel)

#function call: built-in
sharpened = cv.filter2D(img, -1, kernel)

#function call: display images
util.show_sharpen(image_sharpen,sharpened)
```

```python
#Task4 Blur

img = io.imread("images/01.jpg",1)

# Convolve the sharpen kernel (laplacian) and the image
kernel = np.ones((3,3),np.float32)/9

#Call the function my_convolve2d
image_blurred = util.Self_Convolution_Operation(img,kernel)

kernel_size = (3, 3) 
  
# Using cv2.blur() method  
blur = cv.blur(img, kernel_size)

util.show_blur(image_blurred,blur)
```

```python
#Task5
#4 Gaussian Filter and Median Filters

image = cv.imread("images/04.jpg")

gaussian_filter = cv.GaussianBlur(image, (3,3),0)
median_filter = cv.medianBlur(image,3)

util.gaussian_median_show(image, gaussian_filter, median_filter)
```

```python
#Task5
#5 Mesh plot of Gaussian Filter
    
image = cv.imread('images/02.jpg',0) 

gaussian_sig_4 = cv.GaussianBlur(image, (5,5),sigmaX=4)
gaussian_sig_6 = cv.GaussianBlur(image, (5,5),sigmaX=6)
gaussian_sig_8 = cv.GaussianBlur(image, (5,5),sigmaX=8)

util.mesh_plot(gaussian_sig_4, 'Gaussian sigma = 4')
util.mesh_plot(gaussian_sig_6, 'Gaussian sigma = 6')
util.mesh_plot(gaussian_sig_8, 'Gaussian sigma = 8')
```

```python
#Task5
#First Order Derivative of Gaussian
img = cv.imread("Images/04.jpg")

FOD_image = cv.GaussianBlur(image, (5,5),sigmaX=2)
plt.imshow(FOD_image)
util.mesh_plot(FOD_image, 'First Order Derivative')
```

```python
#Task5
#Laplacian of Gaussian

img = cv.imread("Images/05.jpg",0)

gaussian_sig_4 = cv.GaussianBlur(img, (5,5),sigmaX=4)
img_Laplacian = cv.Laplacian(gaussian_sig_4,cv.CV_64F)

plt.imshow(img_Laplacian, cmap=plt.cm.gray)

util.mesh_plot(img_Laplacian, 'Laplacian')
```

```python
#Task 6
#1 Sobel

img = cv.imread("Images/05.jpg",0)

img_SobelX =  cv.Sobel(img,cv.CV_64F,1,0,ksize=5)
img_SobelY = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)

util.sobel_show(img, img_SobelX, img_SobelY)
```

```python
#Task6
#2 Laplacian

img = cv.imread("Images/04.jpg")

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img = cv.GaussianBlur(gray,(3,3),0)
laplacian = cv.Laplacian(img,cv.CV_64F)

util.laplacian_show(gray, img, laplacian)
```
